```python
#scatter plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10,30)        # 1 to 10 , 30 points create krega ye
y=np.sin(x)
plt.plot(x,y,"o",color="red")     # "o" grapgh type ie scattered
```

```python
#2nd way
plt.scatter(x,y)
```

```python
plt.plot(x,y,"-o",color="red")     #line + scattered    '-' for line and 'ok' for scattered
```

```python
plt.plot(x,y,"p",color="red")     # 'p' for hexagonal
```

```python
plt.plot(x,y,"-p",color="red")  
```

```python
x=np.random.randn(100)   #randn standard normal distribution funation
y=np.random.randn(100)
color=np.random.rand(100)
size=1000*np.random.rand(100)
plt.scatter(x,y,c=color,s=size,alpha=0.3)  #alpha for brightness
plt.colorbar()
```

#histogram

```python
data=np.random.randn(1000)
plt.hist(data)    #histogram no need to def x and y axis....
```

```python
plt.hist(data,bins=30,color='pink',alpha=0.4,histtype='stepfilled')   #bin define the spreadness of plot... tani greater the bin greater the point in graph
```

```python
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(0,5,1000)
d=dict(histtype="stepfilled",alpha=0.3,bins=30)
plt.hist(x1,**d)
plt.hist(x2,**d)
plt.hist(x3,**d)
```

```python
x=np.linspace(1,5,100)
y=np.sin(x)
plt.plot(x,y,'--r',label="sin")
y=np.cos(x)
plt.plot(x,y,'-b',label='cos')
plt.legend(loc='upper left',ncol=2)
```

# 2d grapgh

```python
x,y=np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T
plt.hist2d(x,y,bins=30,cmap="Greens")   #cmap for colors
plt.show()
```

# metplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
#plt.plot(x,np.tan(x))
fig=plt.figure()
fig.savefig('lineplot.png')
```

```python
#subplot state base programmng
#state plot
#plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.tan(x))

#object orientedplot
fig,ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="r")
plt.plot(x,np.cos(x),color='chartreuse')
```

```python
x=np.linspace(0,10,1000)
#plt.plot(x,np.sin(x),color="r",linestyle='dotted')
plt.plot(x,np.sin(x),color="r",linestyle=': r')

plt.plot(x,np.cos(x),color='chartreuse',linestyle="dashed")    #solid,dashed,dashdot,dotted
```

```python
#label
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="r",linestyle='dotted')
#plt.plot(x,np.sin(x),color="r",linestyle='.-r')

plt.plot(x,np.cos(x),color='chartreuse',linestyle="dashed")    #solid,dashed,dashdot,dotted

plt.title("grapgh")
plt.xlabel('x')
plt.ylabel('angle(x)')
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="r",linestyle='dotted',label="sin(x)")
#plt.plot(x,np.sin(x),color="r",linestyle=': r')

plt.plot(x,np.cos(x),color='chartreuse',linestyle="dashed",label='cos(x)')    #solid,dashed,dashdot,dotted

plt.title("grapgh")
plt.xlabel('x')
plt.ylabel('angle(x)')
plt.legend()
plt.show()
```

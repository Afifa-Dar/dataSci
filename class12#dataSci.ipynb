```python
t=pd.read_excel("tech2.xlsx")
```

```python
t["NewTech"]=t["Techniques"].str.replace(" ","_")
```

```python
lis=list(t["NewTech"])
print(lis)
```

```python

con=[]
for i in lis:
    tech=sop.find("span",id=i)
    con.append(tech.find_next('p').text)
d={}
c=0
for i in lis:
    d[i]=con[c]
    c+=1
```

```python
d
```

```python
all_data = {"Techniques":lis, "Discription":con}
print(all_data)
```

```python
df = pd.DataFrame(all_data)
df.to_csv("ScrappedTech.csv")
```

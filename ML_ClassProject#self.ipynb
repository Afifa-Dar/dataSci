# Machine Learning

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
```

# split data

```python
x=data[:0:-2]   #not work well
x
```

#array based split

```python
arr=data.values
```

```python
x=arr[:,0:-1]   
y=arr[:,-1]
```

```python
y
```

#series based split

```python
x=data.drop('R',axis=1)
y=data["R"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
y_pred=DT.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print(acc)
```

```python
#we have to apply diff models...so to improve writablity we can......
model=[]
model.append(("DT",DecisionTreeClassifier()))
model.append(("RT",RandomForestClassifier()))
model.append(("KNN",KNeighborsClassifier()))
model.append(("LR",LogisticRegression()))
model.append(("GNB",GaussianNB()))
```

```python
model
```

```python
result=[]
name=[]
for n,m in model:
    name.append(n)
    obj=m
    obj.fit(x_train,y_train)
    y_pred=obj.predict(x_test)
    result.append(accuracy_score(y_test,y_pred)*100)
```

```python
print(name,result)
```

```python
plt.scatter(name,result)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann_model=Sequential([
        Dense(24,input_dim=60,activation="relu"),
        Dense(12, activation='relu'),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')])
ann_model.compile(
    optimizer = keras.optimizers.Adam(),
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
ann_model.summary()
```

# Train model
ANN not take string

```python
#OnneHotEnocding
data["R"].replace(to_replace=["R","M"], value=[0,1], inplace=True)
```

```python
data.tail()
```

```python
x=data.drop("R",axis=1)
y=data["R"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
ann_model.fit(x_train,y_train,epochs=20)          #no of times we train model..by default it is 1
```

```python
#now test accuracy ie evaluate model
#ANN evaluation diff fro ML model evaluation..
#it takes two arg ie x_test y_test and return two values ie loss , accuracy

loss,acc=ann_model.evaluate(x_test,y_test)
```

```python
print(loss)
print(acc)
```

```python
result.append(acc*100)
name.append("ANN")
```

```python
plt.scatter(name,result)
```
